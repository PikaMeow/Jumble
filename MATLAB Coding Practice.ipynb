## 数据的可视化

1.在同一坐标系中画出$y=x^2$,$y=cos$ $x$的图像

```python
%matplotlib inline

from pylab import * 

x=linspace(-pi,pi,256)
y1,y2=x**2,cos(x)
xlim(-4,4)
xticks([-pi,-pi/2,0,pi/2,pi],
      [r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])
ylim(-2,4)
ax=gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plot(x,y1,label="$x^2$")
plot(x,y2,label="$cos x$")
legend(loc='upper right')
show()
```

3.绘制二元函数$f(x,y)=sin(xy)x^2y^3$在$[-4,4]\times [-4,4]$上的表面图、等高线图、伪彩图以及着色表面图

```python
%matplotlib inline

import numpy as np
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

x=np.linspace(-4,4,256)
y=np.linspace(-4,4,256)
X,Y=np.meshgrid(x,y)
Z=(np.sin(X*Y))*(X**2)*(Y**3)
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=cm.coolwarm,\
                linewidth=0,antialiased=False)
plt.show()
```

## 符号运算 

1.求定积分$\int_a^bx^4sin^2(x)dx$,并利用化简函数进行化简

```python
from sympy import *
x,y,a,b=symbols('x y a b')
f=integrate(x**4*sin(x)**2,[x,a,b])
simplify(f)
```

2.求三元函数$u(x,y,z)=sin(xy)e^{x^2ysin(z)}$   的偏微分$\frac{\partial ^5u(x,y,z)}{\partial x^2\partial y^2\partial z}\arrowvert_{(1,2,1)}$

```python
from sympy import *
x,y,z=symbols('x y z')
u=Function('u')
u=(sin(x*y))*(E**(x**2*y*sin(z)))
du=diff(diff(diff(u,x,2),y,2),z)
ddu=du.subs({x:1,y:2,z:1})
simplify(ddu)
```
